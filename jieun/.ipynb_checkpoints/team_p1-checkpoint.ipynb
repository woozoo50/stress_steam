## 2023년 지역별 고혈압 그래프

```python
import pandas as pd
import matplotlib.pyplot as plt

# 엑셀 파일 읽기
file_path = './datas/go.xlsx'
df = pd.read_excel(file_path)

# 필요한 데이터 추출
df_plot = df[['시군구별(1)', '2023']]

# 그래프 그리기
plt.figure(figsize=(10, 6))

# colors = ['#FF5733', '#33FF57', '#3357FF', '#FF33A8', '#FFBD33', '#33FFF5', '#8D33FF', '#FF3333', '#33FF33', '#3333FF']
plt.bar(df_plot['시군구별(1)'], df_plot['2023'])

# 제목 및 레이블 설정
plt.rc('font', family='Malgun Gothic')
plt.title('2023년 지역별 고혈압')
plt.xlabel('')
plt.ylabel('2023년 값')

# 그래프 레이아웃 조정 및 표시
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## 서울특별시 연도별 스트레스 및 고혈압 수치 변화 그래프

```python
import pandas as pd
import matplotlib.pyplot as plt

# 엑셀 파일 읽기
file_path_stress = './datas/stress.xlsx'
file_path_hypertension = './datas/go.xlsx'

# 데이터프레임 읽기
df_stress = pd.read_excel(file_path_stress)
df_hypertension = pd.read_excel(file_path_hypertension)

# 서울특별시의 데이터 추출
seoul_stress = df_stress[df_stress['시군구별(1)'] == '서울특별시']
seoul_hypertension = df_hypertension[df_hypertension['시군구별(1)'] == '서울특별시']

# Figure와 두 개의 축(ax) 생성
fig, ax1 = plt.subplots(figsize=(10, 6))

# 첫 번째 축: 스트레스 변화 (막대 그래프)
ax1.bar(seoul_stress.columns[1:], seoul_stress.iloc[0, 1:], color='skyblue', label='스트레스 변화')
ax1.set_ylabel('스트레스 수치')
ax1.set_ylim(20, 31)
ax1.set_yticks([23, 24, 25, 26, 27, 28, 29, 30, 31])
ax1.tick_params(axis='y')

# 두 번째 축: 고혈압 수치 변화 (선 그래프)
ax2 = ax1.twinx()  # 두 번째 y축을 공유
ax2.plot(seoul_hypertension.columns[1:], seoul_hypertension.iloc[0, 1:], marker='o', color='orange', linestyle='-', label='고혈압 수치 변화')
ax2.set_ylabel('고혈압 수치')
ax2.set_ylim(20, 28)
ax2.set_yticks([20, 21, 22, 23, 24, 25, 26, 27, 28])
ax2.tick_params(axis='y')

# 제목 설정
plt.title('서울특별시 연도별 스트레스 및 고혈압 수치 변화')

# x축 레이블 설정 및 그래프 표시
plt.xticks(rotation=45)
fig.tight_layout()
plt.grid(True, axis='y', linestyle='--', linewidth=0.7)
plt.show()
```

## 서울특별시 스트레스와 고혈압 상관계수 그래프

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 엑셀 파일 읽기
file_path_stress = './datas/stress.xlsx'
file_path_hypertension = './datas/go.xlsx'

# 데이터프레임 읽기
df_stress = pd.read_excel(file_path_stress)
df_hypertension = pd.read_excel(file_path_hypertension)

# 서울특별시의 데이터 추출
seoul_stress = df_stress[df_stress['시군구별(1)'] == '서울특별시']
seoul_hypertension = df_hypertension[df_hypertension['시군구별(1)'] == '서울특별시']

# 연도별 스트레스와 고혈압 수치 데이터 결합
combined_data = pd.DataFrame({
    '연도': seoul_stress.columns[1:],
    '스트레스': seoul_stress.iloc[0, 1:].values,
    '고혈압': seoul_hypertension.iloc[0, 1:].values
})

# 상관계수 계산
correlation = combined_data[['스트레스', '고혈압']].corr()

# 상관계수 히트맵 그리기
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('서울특별시 스트레스와 고혈압 수치 상관계수')
plt.show()
```

## 대한민국 연도별 스트레스 수치 변화

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# CSV 파일 로드
file_path = 'team_data/stress1.csv'  # 파일 경로를 실제 경로로 수정하세요.
data = pd.read_csv(file_path)

# '시군구별(1)' 열을 인덱스로 설정
data.set_index('시군구별(1)', inplace=True)

# 전국 평균 계산 (모든 시군구의 평균을 계산)
national_average = data.mean(axis=0)

# 그래프 그리기
plt.figure(figsize=(12, 8))

# 흰색 배경 설정
plt.gca().set_facecolor('white')

# 막대 그래프 (바 그래프)
bars = plt.bar(national_average.index, national_average.values, color='lightblue', label='스트레스 수치')

# 꺾은선 그래프 (라인 그래프)
plt.plot(national_average.index, national_average.values, color='blue', marker='o', linestyle='--', label='수치값')

# 막대 그래프 위에 수치 표시
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, round(yval, 1), ha='center', va='bottom', fontsize=12)

# 그래프 제목 및 축 레이블 설정
plt.rc('font', family='Malgun Gothic')
plt.title('대한민국 연도별 스트레스 수치 변화', fontsize=18, fontweight='bold', pad=20)
plt.xlabel('연도', fontsize=14, labelpad=10)
plt.ylabel('스트레스 수치 (%)', fontsize=14, labelpad=10)

# x축 눈금 레이블 회전 및 글꼴 크기 조정
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)

# 가로선만 있는 그리드 추가
plt.grid(axis='y', linestyle='--', color='gray')

# 범례 추가
plt.legend(loc='upper right')

# 그래프 테두리와 레이아웃 조정
plt.tight_layout()

# 그래프 출력
plt.show()
```
