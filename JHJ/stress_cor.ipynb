```python
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
import csv

plt.rc('font', family='Malgun Gothic')

#%matplotlib inline 
# notebook을 실행한 브라우저에서 바로 그래프를 볼 수 있게 하는 기능

cause_list = []
stress_list=[]

with open("stress/smoke.csv") as f:
    data = csv.reader(f)
    data=list(data)
    for i in data:
        for z in i[1:]:
            cause_list.append(float(z))
                  
with open('stress/stress.csv') as f2:
    data = csv.reader(f2)
    next(data)
    data=list(data)
    for i in data:
        for z in i[1:]:
            stress_list.append(float(z))

df = pd.DataFrame()
df['흡연율'] = cause_list
df['스트레스'] = stress_list

plt.scatter(df['스트레스'], df['흡연율']) 
plt.show() #그래프 표시

correlation = df.corr()
print(correlation)

sns.lmplot(x='스트레스', y='흡연율', data=df, height=6, aspect=2)
plt.suptitle('흡연율과 스트레스 수준 간의 관계', y=1.02)
plt.show()
```

```python
import pandas as pd
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

warnings.simplefilter('ignore')
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False

categories = [
    '   비경제\n   활동',
    '  경제\n  활동', 
    '여가시간', 
    '흡연율', 
    '고혈압   ', 
    '실업자   ', 
    '우울증', 
    '문화체육시설', 
    '  취업자',
]

categories = [*categories, categories[0]] # 배열의 0번 인덱스를 맨뒤에 1개 추가 ( 0 > 1 > 2 ... i > 0  ) 

grade1 = [ 0.255896, 0.036931, 0.509334, 0.398441, 0.324835, 0.31505, 0.289361, 0.28153, 0.260096]
grade1 = [*grade1, grade1[0]] # 배열의 0번 인덱스를 맨뒤에 1개 추가 ( 0 > 1 > 2 ... i > 0  ) 

Best1 = [0.509334, 0.398441] # 여가시간에 점을 찍기 위한 좌표
Best2 = [1.3962634,  2.0943951] # 흡연율에 점을 찍기 위한 좌표

#linspace : start에서 stop까지 num개 간격으로 배열 만듬
label_loc = np.linspace(start=0, stop=2*np.pi, num=len(grade1))  # 그래프 선을 긋기 위한 각도값 , 0에서 360도까지를 , 배열수 간격으로  
print(label_loc/3.144*180)

plt.figure(figsize=(10, 10))
ax = plt.subplot(polar=True)  # 극좌표계

plt.xticks(label_loc, labels=categories, fontsize=12)
# plot(각도, 거리)
ax.plot(label_loc, grade1, label='상관계수', linestyle='dashed', color='skyblue', alpha=0.4)  # alpha는 투명도
ax.fill(label_loc, grade1, color='skyblue', alpha=0.25)

plt.scatter(Best2, Best1, color='blue', zorder=3)
plt.text(1.3562634, 0.503, "0.51", color='blue', zorder=3, fontsize=12)
plt.text(2.1, 0.42, "0.4", color='blue', zorder=3, fontsize=12)

ax.legend()  # 범례 추가
plt.show()
```
