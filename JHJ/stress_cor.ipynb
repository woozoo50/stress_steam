```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
import csv

plt.rc('font', family='Malgun Gothic')

#%matplotlib inline 
# notebook을 실행한 브라우저에서 바로 그래프를 볼 수 있게 하는 기능

cause_list = []
stress_list=[]

with open("stress/smoke.csv") as f:
    data = csv.reader(f)
    data=list(data)
    print(data)
    for i in data:
        for z in i[1:]:
            cause_list.append(float(z))
                  
with open('stress/stress.csv') as f2:
    data = csv.reader(f2)
    next(data)
    data=list(data)
    print(data)
    for i in data:
        for z in i[1:]:
            stress_list.append(float(z))

df = pd.DataFrame()
df['흡연율'] = cause_list
df['스트레스'] = stress_list

df

plt.scatter(df['스트레스'], df['흡연율']) 
plt.show() #그래프 표시

correlation = df.corr()
print(correlation)

sns.lmplot(x='스트레스', y='흡연율', data=df, height=6, aspect=2)
plt.suptitle('흡연율과 스트레스 수준 간의 관계', y=1.02)
plt.show()
```
